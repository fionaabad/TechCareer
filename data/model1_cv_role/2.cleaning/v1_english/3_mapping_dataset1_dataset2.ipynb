```python
import pandas as pd
```

```python
d1 = pd.read_csv("resumes_dataset1_clean.csv")
d2 = pd.read_csv("resumes_dataset2_clean.csv")
```

```python
print("Dataset 1 (Avishek):", d1.shape)
print("Dataset 2 (Jithin):", d2.shape)

print("\nRoles dataset 1:")
print(d1['role_label_final'].value_counts())

print("\nRoles dataset 2:")
print(d2['role_label_final'].value_counts())
```

```python
# Unir todos los roles crudos de ambos datasets
roles_d1 = d1[['role_raw']].assign(source_dataset='dataset1_avishek')
roles_d2 = d2[['role_raw']].assign(source_dataset='dataset2_jithin')

all_roles = (
    pd.concat([roles_d1, roles_d2], ignore_index=True)
    .drop_duplicates()
    .sort_values(['source_dataset', 'role_raw'])
)

all_roles
```

```python
all_roles.to_csv("role_mapping_manual.csv", index=False)
print("Guardado role_mapping_manual.csv")
```

```python
import pandas as pd

# Cargar el mapping que acabas de guardar
mapping = pd.read_csv("role_mapping_manual.csv")
print(mapping)

# Diccionario: role_raw -> nombre canónico para el modelo
canonical_map = {
    "Python_Developer": "python_developer",
    "Python Developer": "python_developer",
    "Front_End_Developer": "frontend_developer",
    "Frontend Developer": "frontend_developer",
    "Java_Developer": "java_developer",
    "Web_Developer": "web_developer",
    "Database_Administrator": "database_administrator",
    "Security_Analyst": "security_analyst",
    "Systems_Administrator": "systems_administrator",
    "Network_Administrator": "network_administrator",
    "Project_manager": "project_manager",
    "Software_Developer": "software_developer",
    "Backend Developer": "backend_developer",
    "Cloud Engineer": "cloud_engineer",
    "Data Scientist": "data_scientist",
    "Full Stack Developer": "fullstack_developer",
    "Machine Learning Engineer": "ml_engineer",
    "Mobile App Developer (iOS/Android)": "mobile_app_developer",
}

# Crear la nueva columna usando el diccionario
mapping["role_label_final"] = mapping["role_raw"].map(canonical_map)

# Comprobar si algún role_raw se ha quedado sin mapeo
print("Roles sin mapeo en canonical_map:",
      mapping["role_label_final"].isna().sum())

mapping
```

```python
mapping.to_csv("role_mapping_manual.csv", index=False)
print("Mapping actualizado guardado.")
```
